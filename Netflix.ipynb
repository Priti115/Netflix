#### which Individual Country has the hieghest number of TV Show ?

### First method

```python
data[data['Category'] == 'TV Show']['Country'].value_counts().idxmax()
```

```python
data[data['Category'] == 'TV Show']['Country'].value_counts().max()
```

### second method

```python
tv_shows = data[data['Category'] == 'TV Show'] # here find out the all TV Show in the Category 
```

```python
country_counts = tv_shows['Country'].value_counts()   # here we have found the total country
```

```python
highest_tv_shows_country = country_counts.idxmax()
highest_tv_shows_count = country_counts.max()
```
